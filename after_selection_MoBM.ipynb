```python
import numpy as np
import pandas as pd

import pickle
```

```python
## load the HOMFLYPT data
homfly_df = pd.read_csv('data/HomflyPt_upto_15_MIRRORS.csv')
homfly_df
```

```python
MOBM = 'NA_50_40' # the id of the MoBM we are interested in
```

```python
# load the MoBM graph
with open('pkl/mobm_{}.pkl'.format(MOBM), "rb") as f:
    G = pickle.load(f)

print('{} nodes loaded'.format(len(G.nodes)))
```

```python
# load the nodes that were selected in the GUI
with open('pkl/mobm_{}_selected_nodes.pkl'.format(MOBM), "rb") as f:
    selected_nodes = pickle.load(f)

print(len(selected_nodes))
```

```python
# get the list of points covered by the selected nodes
points_id = list(set([p for node in selected_nodes for p in G.nodes[node]['points_covered']]))
len(points_id)
```

```python
# slice through the df
subset_df = homfly_df.iloc[points_id].copy()
```

```python
homfly_df.iloc[points_id].signature.value_counts()
```

```python
homfly_df.iloc[points_id].s_invariant.value_counts()
```
