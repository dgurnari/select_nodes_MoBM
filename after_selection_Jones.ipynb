```python
## load the knot_info df

knot_info_df = pd.read_csv('data/coloring_df.csv')
knot_info_df['s_signature_diff'] = np.abs(knot_info_df['signature'].abs() -
                                          knot_info_df['s_invariant'].abs())
knot_info_df
```

```python
knot_info_df.s_signature_diff.value_counts()
```

```python
# load the Jones BM graph
with open('pkl/jones_NA_50.pkl', "rb") as f:
    jones_bm = pickle.load(f)

print('{} nodes loaded'.format(len(jones_bm.Graph.nodes)))
```

```python
# load the nodes that were selected in the GUI
with open('pkl/jones_NA_50_selected_nodes.pkl', "rb") as f:
    selected_nodes = pickle.load(f)
    selected_nodes = [int(n) for n in selected_nodes]

print(len(selected_nodes))
```

```python
# get the list of points covered by the selected nodes
points_id = list(set([p for node in selected_nodes for p in jones_bm.Graph.nodes[node]['points covered']]))
len(points_id)
```

```python
# slice through the df
subset_df = knot_info_df.iloc[points_id].copy()
subset_df
```

```python
subset_df.signature.value_counts()
```
